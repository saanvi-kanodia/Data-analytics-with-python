```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
from statsmodels.formula.api import ols 
```

```python
df = pd.read_excel('Twoway.xlsx')
df
```

```python
formula = 'Value ~C(college)+C(prep_pro)+C(college):C(prep_pro)'
#                  Factor B, Factor A, Interaction
model = ols(formula, df).fit()
anova_table = sm.stats.anova_lm(model, type =2)
anova_table
```

```python
# we accept H0 for indifference in effect caused by preparation program and interaction
# we reject H0 for indifference in effect caused by college
# college background has effect on results 
# here the college is significant effect. 
```
