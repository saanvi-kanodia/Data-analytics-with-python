```python
import pandas as pd
import numpy as np
import scipy
import statsmodels.api as sm
from statsmodels.formula.api import ols
```

#### Q1)

```python
df = pd.read_excel('./RBD.xlsx')
df
```

```python
dfnew = pd.melt(df.reset_index(),id_vars=['index'],value_vars=['System A', 'System B', 'System C'])
dfnew.columns=['index','treatment','value']
```

```python
model = ols('value ~ C(treatment)', data = dfnew).fit()
anova_table = sm.stats.anova_lm(model,type=1)
anova_table
#type 1 for one way anova
```

```python
#accept null hypothesis ( as p falue is greater that alpha(0.05))
```

```python
#now we will do Blocking to remove variance within each treatment(MSE) 
#so that the variance between treatments is more visible 
#so true F is visible
```

```python
data1 = pd.melt(df.reset_index(),id_vars=["index"], value_vars=["System A", "System B", "System C"])
data1.columns = ['blocks', 'treatments', 'value'] 
```

```python
model = ols('value ~ C(blocks)+C(treatments)', data = data1).fit()
anova_table = sm.stats.anova_lm(model, type=1)
anova_table
```

```python
#PR value for C(treatments) is less than alpha(0.05)
#so we reject null hypothesis 
```

#### Q2)
test in difference in mean using ANOVA with alpha = 0.01
first we will do normal F test, than RBD

```python
data2 = pd.read_excel('rbd2.xlsx')
data2
```

```python
d = pd.melt(data2.reset_index(), id_vars=['index'], value_vars=['chem1','chem2','chem3','chem4'])
d.columns = ['index','treatment', 'values']
d
model = ols('values ~ C(treatment)', data=d).fit()
anova_table=sm.stats.anova_lm(model,type=1)
anova_table
```

```python
#since p<alpha, we reject it
```

```python
data3 = pd.melt(data2.reset_index(),id_vars=['index'], value_vars=['chem1','chem2','chem3','chem4'])
data3.columns = ['blocks', 'treatments','values']
data3
```

```python
model = ols('values ~ C(blocks) + C(treatments)',data = data3).fit()
anova_table = sm.stats.anova_lm(model, type=1)
anova_table
```

```python
#the SSE derived above (sum of squares of residual) 
#is the true error
#F is so great, the chance of rejections is so high
#F << F(alpha 0.01), we reject null hypothesis
#significant difference in chemical effects
```
