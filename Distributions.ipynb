#### HyperGeometric Distribution
finite population with replacement
* hypergeom.sf(x-1, N, n, A)
* sf - survival function
* x -> sample success event size
* N -> population size
* n -> sample size
* A -> population sucess size

```python
from scipy.stats import hypergeom
```

```python
pval = hypergeom.sf(0,18,3,12)
# sf(x-1, N, n, A) ; x=1  question stated 1 or more so we are basically subtracting from one like we usually do
```
